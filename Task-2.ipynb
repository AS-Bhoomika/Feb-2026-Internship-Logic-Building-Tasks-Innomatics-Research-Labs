Problem 1: Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")

words = sentence.split()
unique_words = set(words)

print(sentence)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
# Input: Dictionary of employees and their salaries
employees = {
    "Ravi": 25700,
    "Anita": 629000,
    "Kiran": 586000
}

# Logic: Find the employee with the maximum salary
# NOW We are useing max() with a key argument to find the key corresponding to the highest value
highest_earning_employee = max(employees, key=employees.get)
highest_salary = employees[highest_earning_employee]

# Output
print(f"Highest Salary: {highest_earning_employee} - {highest_salary}")
```

Problem 3: Find Maximum and Minimum Values

```python
numbers = [14, 21, 69, 25, 19]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [200, 1400, 898, 1200, 700]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

Problem 7: Count Character Frequency

```python
string = "learning python programming"

char_count = {}

for char in string:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

Problem 9: Check if a Key Exists in a Dictionary

```python
employees = {
    "Rahul": 95000,
    "Abhi": 69000
}

key_to_check = "Rahul"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
