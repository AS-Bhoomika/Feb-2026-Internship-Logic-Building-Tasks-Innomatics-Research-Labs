## 2. Pass / Fail Analyzer

Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks ≥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly

```python
marks=[45,78,90,33,60]
passCount=0
failCount=0
for i in marks:
    if i>=50:
        passCount+=1
        print("Marks=",i," Pass")
    else:
        failCount+=1
        print("Marks=",i," Fail")
print("Number of Total pass students are ",passCount)
print("Number of Total failed students are ",failCount)
```

## 3. Simple Data Cleaner

Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list

```python
names = [" Alice ", "bob", " CHARLIE "]
newList=[]
for name in names:
    #To convert in lower case without spaces
    name=name.strip().lower()
    # To Append to the newList
    newList.append(name)
print(newList)
```

## 4. Message Length Analyzer

Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:
Print the length of each message
Flag messages longer than 10 characters

```python
messages = ["Hi", "coders Welcome to my World", "OK"]
for msg in messages:
    length=len(msg)
    if length>10:
        flag="[Long Msg]"
    else:
        flag="[OK]"
    print(f"Message: {msg} | {length} {flag}")
```

## 5. Error Message Detector

Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
errors=0
for err in logs:
    if "ERROR" in err:
        errors+=1
print("The number of ERROR entries are",errors)
```
